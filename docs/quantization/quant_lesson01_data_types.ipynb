## Floating Points

Three components in floating point:
- Sign (1 bit): The leftmost bit. `0` for positive, `1` for negative.
- Exponent, or range: impact the representable range of the number
- Mantissa or Fraction (precision): impact on the precision of the number

FP32, BF16, FP16, FP8 are floating point format with a specific number of bits for exponent and the fraction.

## Code Example to Decode Binary String to Decimal FP32 and Vice Versa

### From Binary String to FP32 Value

An FP32 (single-precision floating-point) number is represented according to the IEEE 754 standard and is composed of three parts:
- Sign bit: The first bit, which determines if the number is positive (0) or negative (1).
- Exponent: The next 8 bits, which represent the power of 2 by which the mantissa is multiplied.
- Mantissa (or fraction): The remaining 23 bits, which represent the significant digits of the number.

The full formula for the decimal value is:

$$(-1)^{\text{sign}}\times (1+\text{mantissa})\times 2^{(\text{exponent}-\text{bias})}$$

Here is a step-by-step breakdown of how to decode an FP32 number.

**1. Extract the components**

Take the 32-bit binary representation of the floating-point number and divide it into its three parts.

Example: A number represented as `01000000010000000000000000000000`
- Sign bit: `0` (indicates a positive number)
- Exponent bits: `10000000`
- Mantissa bits: `10000000000000000000000`

**2. Decode the sign**

The first bit, the sign bit, is the easiest to decode.

- If the sign bit is 0, the number is positive.
- If the sign bit is 1, the number is negative.

In the example, the sign bit is 0, so the number is positive.

**3. Decode the exponent**

The 8-bit exponent field is a biased representation. For FP32, the bias is `127`. To get the actual exponent value, you must convert the binary exponent to decimal and then subtract the bias.

Steps:
- Convert the exponent bits to a decimal number. In the example, the exponent bits are `10000000`.
$$ 1\times 2^{7}+0\times 2^{6}+0\times 2^{5}+0\times 2^{4}+0\times 2^{3}+0\times 2^{2}+0\times 2^{1}+0\times 2^{0}=128$$

- Subtract the bias (127) to get the true exponent, as 128-127=1

**4. Decode the mantissa**

The mantissa is the fractional part of the number. The IEEE 754 standard assumes a leading 1 that is not explicitly stored for normalized numbers.

Steps:
- Add the implicit leading 1 to the beginning of the mantissa bits, separated by a decimal point. In the example, the mantissa bits are `10000000000000000000000`. So, the full mantissa is `1.10000000000000000000000`.
- Convert the mantissa from binary to decimal. The places after the decimal point represent negative powers of 2.

$$1\times 2^{0}+1\times 2^{-1}+0\times 2^{-2}+\dots =1+0.5=1.5$$

**5. Combine the parts**

Now, assemble the final decimal value using the formula:

$$(-1)^{\text{sign}}\times (\text{mantissa})\times 2^{\text{exponent}}$$

Applying the formula to the example:
- Sign: Positive (-1^0)
- Mantissa: 1.5
- Exponent: 1

Final calculation:
$$+1\times 1.5\times 2^{1}=3.0$$

````python
import struct
# The 'f' format code in struct handles all
# the IEEE 754 calculations for you.
def fp32_binary_to_decimal(binary_string):
    # Convert the 32-bit binary string to an integer
    integer_value = int(binary_string, 2)

    # Pack the integer into 4 bytes (for FP32)
    packed_bytes = struct.pack('>I', integer_value) # '>I' for big-endian, unsigned int

    # Unpack the bytes as a float
    decimal_value = struct.unpack('>f', packed_bytes)[0] # '>f' for big-endian float

    return decimal_value
````

````python
# Example usage with the same binary string
binary_string = '01000000010000000000000000000000'
result = fp32_binary_to_decimal(binary_string)
print(result) # Output: 3.0
````

The `'f'` format code in struct handles all the IEEE 754 calculations for us.

It is possible to decode an FP32 binary string to a decimal value manually, without using the struct module.

The process requires a function that explicitly implements the IEEE 754 standard by breaking the 32-bit string into its three components: sign, exponent, and mantissa.

````python
def fp32_binary_to_decimal_manual(binary_string):
    """
    Manually converts a 32-bit FP32 binary string to a decimal floating-point number.

    Args:
        binary_string (str): A 32-character binary string representing an FP32 value.

    Returns:
        float: The decimal floating-point value.
    """
    # Check for non-standard length
    if len(binary_string) != 32 or not all(c in '01' for c in binary_string):
        raise ValueError("Input must be a 32-character binary string.")

    # 1. Extract the components based on IEEE 754 standard
    sign_bit = binary_string[0]
    exponent_bits = binary_string[1:9]
    mantissa_bits = binary_string[9:]

    # 2. Decode the sign
    sign = -1.0 if sign_bit == '1' else 1.0

    # 3. Decode the exponent
    # Convert the 8-bit binary exponent to an integer
    exponent_val = int(exponent_bits, 2)
    # The bias for FP32 is 127
    exponent = exponent_val - 127

    # 4. Decode the mantissa
    mantissa = 0.0
    for i, bit in enumerate(mantissa_bits):
        if bit == '1':
            # The mantissa bits represent negative powers of 2 (1*2^-1, 1*2^-2, etc.)
            mantissa += 2 ** (-(i + 1))

    # 5. Handle special cases based on the exponent
    if exponent_val == 0:
        if mantissa == 0:
            return sign * 0.0  # Zero (positive or negative)
        else:
            # Denormalized number: 2^(exponent_min) * (0 + mantissa)
            # In FP32, exponent_min is -126
            return sign * (mantissa) * (2**-126)
    elif exponent_val == 255:
        if mantissa == 0:
            return sign * float('inf')  # Infinity (positive or negative)
        else:
            return float('nan')  # Not a number
    else:
        # Normalized number: 2^(exponent) * (1 + mantissa)
        return sign * (1.0 + mantissa) * (2**exponent)
````

````python
# --- Example Usage ---
# Example 1: 3.0
binary_str_1 = '01000000010000000000000000000000'
result_1 = fp32_binary_to_decimal_manual(binary_str_1)
print(f"Binary: {binary_str_1}")
print(f"Decimal: {result_1}")  # Expected: 3.0
````

### From FP32 Value to Binary String

````python
def decimal_float_to_fp32_binary_manual(float_value):
    """
    Converts a Python float to its FP32 binary representation manually,
    without using the struct module.

    Args:
        float_value (float): The float value to convert.

    Returns:
        str: A 32-character binary string representing the FP32 value.
    """
    if float_value == 0.0:
        return '0' * 32
    if float_value < 0:
        sign = '1'
        float_value = -float_value
    else:
        sign = '0'

    # Handle special cases (e.g., infinity, NaN)
    # These are not covered by this manual function and require struct.

    # 1. Convert integer and fractional parts to binary
    integer_part = int(float_value)
    fractional_part = float_value - integer_part

    # Convert integer part to binary
    if integer_part == 0:
        binary_integer = ''
    else:
        binary_integer = bin(integer_part)[2:]

    # Convert fractional part to binary
    binary_fractional = ''
    while fractional_part > 0 and len(binary_fractional) < 24:
        fractional_part *= 2
        bit = int(fractional_part)
        binary_fractional += str(bit)
        fractional_part -= bit

    binary_value = binary_integer + '.' + binary_fractional

    # 2. Normalize the binary number
    if integer_part > 0:
        # For numbers >= 1.0
        # Example: 10.75 -> 1010.11 -> 1.01011 * 2^3
        exponent_val = len(binary_integer) - 1
        mantissa_str = binary_integer[1:] + binary_fractional
    else:
        # For numbers between 0 and 1
        # Example: 0.75 -> 0.11 -> 1.1 * 2^-1
        if fractional_part == 0:
             return '0' * 32 # Denormalized zero

        first_one_index = binary_fractional.find('1')
        exponent_val = -(first_one_index + 1)
        mantissa_str = binary_fractional[first_one_index + 1:]

    # 3. Calculate the biased exponent
    biased_exponent = exponent_val + 127
    binary_exponent = bin(biased_exponent)[2:].zfill(8)

    # 4. Construct the mantissa (23 bits)
    mantissa = mantissa_str[:23].ljust(23, '0')

    # 5. Assemble and return the full FP32 binary string
    return sign + binary_exponent + mantissa
````

````python
# --- Example Usage ---
# Example 1: 3.0
input_float = 3.0
fp32_binary = decimal_float_to_fp32_binary_manual(input_float)
res_float = fp32_binary_to_decimal_manual(fp32_binary)
print(f"Input: {input_float}")
print(f"FP32 Binary: {fp32_binary}")
print(f"Result: {res_float}")

print("-" * 30)
# Example 2: 10.75
input_float = 10.75
fp32_binary = decimal_float_to_fp32_binary_manual(input_float)
res_float = fp32_binary_to_decimal_manual(fp32_binary)
print(f"Input: {input_float}")
print(f"FP32 Binary: {fp32_binary}")
print(f"Result: {res_float}")

print("-" * 30)
# Example 2: -0.1
input_float = -0.1
fp32_binary = decimal_float_to_fp32_binary_manual(input_float)
res_float = fp32_binary_to_decimal_manual(fp32_binary)
print(f"Input: {input_float}")
print(f"FP32 Binary: {fp32_binary}")
print(f"Result: {res_float}")
````

Using Python's built-in struct module is the most reliable way to convert a standard Python float (which is 64-bit) into a 32-bit (FP32) binary string. This method correctly handles the IEEE 754 floating-point standard used by computers.

````python
import struct

def decimal_float_to_fp32_binary(float_value):
    """
    Converts a Python float to its 32-bit (FP32) IEEE 754 binary string representation.

    Args:
        float_value: The float value to convert.

    Returns:
        A 32-character binary string representing the FP32 value.
    """
    # Pack the Python float (64-bit) into a 4-byte (32-bit) structure
    # '>f' specifies big-endian format for a single-precision float
    packed_bytes = struct.pack('>f', float_value)

    # Unpack the 4 bytes as a 32-bit unsigned integer
    # '>I' specifies big-endian format for a standard unsigned integer
    unpacked_int, = struct.unpack('>I', packed_bytes)

    # Format the integer as a 32-character binary string with leading zeros
    return f'{unpacked_int:032b}'
````

````python
# --- Example Usage ---
# Example 1: 3.0
input_float = 3.0
fp32_binary = decimal_float_to_fp32_binary(input_float)
res_float = fp32_binary_to_decimal(fp32_binary)
print(f"Input: {input_float}")
print(f"FP32 Binary: {fp32_binary}")
print(f"Result: {res_float}")

print("-" * 30)
# Example 2: 10.75
input_float = 10.75
fp32_binary = decimal_float_to_fp32_binary(input_float)
res_float = fp32_binary_to_decimal(fp32_binary)
print(f"Input: {input_float}")
print(f"FP32 Binary: {fp32_binary}")
print(f"Result: {res_float}")

print("-" * 30)
# Example 2: -0.1
input_float = -0.1
fp32_binary = decimal_float_to_fp32_binary(input_float)
res_float = fp32_binary_to_decimal(fp32_binary)
print(f"Input: {input_float}")
print(f"FP32 Binary: {fp32_binary}")
print(f"Result: {res_float}")
````
